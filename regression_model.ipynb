```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('/content/Data_Train.xlsx')
data.head()
```

```python
data.info()
```

## Data Cleaning

```python
data['Day'],data['Month'],data['Year']=data['Date_of_Journey'].str.split('/',3).str  
data['Day'],data['Month'],data['Year']=data['Day'].astype(int),data['Month'].astype(int),data['Year'].astype(int)
data.head()
```

```python
data.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['Dep_Hr'],data['Dep_Min']=data['Dep_Time'].str.split(':',2).str
data['Dep_Hr'],data['Dep_Min']=data['Dep_Hr'].astype(int),data['Dep_Min'].astype(int)
data.drop(['Dep_Time'],axis=1,inplace=True)
data.head()
```

```python
data['Arrival_Time'],_=data['Arrival_Time'].str.split(' ',1).str
data.head()
```

```python
data['Arr_Hr'],data['Arr_Min']=data['Arrival_Time'].str.split(':',2).str
data['Arr_Hr'],data['Arr_Min']=data['Arr_Hr'].astype(int),data['Arr_Min'].astype(int)
data.drop(['Arrival_Time'],axis=1,inplace=True)
data.head()
```

```python
data['Duration']=data['Duration'].str.replace('h ',":").str.replace('m',"").str.replace('h',':00')
data.head()
```

```python
data['dur_Hr'],data['dur_Min']=data['Duration'].str.split(':',2).str
```

```python
data['dur_Hr'].unique()
```

```python
data['dur_Min'].fillna(0,inplace=True)
data['dur_Min'].unique()
```

```python
data['dur_Hr'],data['dur_Min']=data['dur_Hr'].astype(int),data['dur_Min'].astype(int)
data.drop(['Duration'],axis=1,inplace=True)
data.head()
```

```python
data['Duration']=(data['dur_Min']/60 + data['dur_Hr'])
data.drop(['dur_Hr'	,'dur_Min'],axis=1,inplace=True)
data.head()
```

```python
data['Total_Stops'].unique()
```

```python
data['Total_Stops']=data['Total_Stops'].str.replace(' stops',"").str.replace(' stop',"").str.replace("non-stop",'0')
data['Total_Stops'].unique()
```

```python
data.dropna(inplace=True)
```

```python
data['Total_Stops']=data['Total_Stops'].astype('int')
data.head()
```

```python
data['Additional_Info'].unique()
```

```python
data['Additional_Info']=data['Additional_Info'].str.replace('No info','No Info')
data['Additional_Info'].value_counts()
```

```python
data.drop(['Route'],axis=1,inplace=True)
data.head()
```

```python
data=pd.get_dummies(data,drop_first=True)
data.head()
```

```python
data.isna().sum()
```

```python
X = data.drop("Price", axis=1)
y = data['Price']
```

# **split data to train  &  test**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# **DecisonTree model**

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(random_state = 0) 
```

```python
regressor.fit(X_train,y_train)
```

```python
regressor.score(X_train,y_train)*100
```

```python
regressor.fit(X_test,y_test)
```

```python
regressor.score(X_test,y_test)
```

```python
from sklearn.tree import export_graphviz 
```

```python
export_graphviz(regressor, out_file ='tree.dot',
               feature_names =X.columns)
```

## **random forest model**


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg=RandomForestRegressor(n_estimators=100,random_state=0)
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_train,y_train)
```

```python
reg.fit(X_test,y_test)
```

```python
reg.score(X_test,y_test)
```
